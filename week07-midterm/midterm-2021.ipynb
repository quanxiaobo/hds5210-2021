## Part 1: CHA2DS2-VASc

This scoring mechanism for Atrial Fibrillation Stroke uses 7 inputs:
* Age (Number)
* Sex (Male / Female)
* CHF History (True / False)
* Hypertension History (True / False)
* Stroke History (True / False)
* Vascular Disease History (True / False)
* Diabetes History (True / False)

Fill out the function below with logic to calculate the numeric risk score for teh given input.

Be sure to provide meaningful documentation and at least two test cases in your documentation.  Also make sure your code satisfies the test cases provided in the assert statements.

````python
def cha2ds2_vasc(age, sex, chf, hypertension, stroke, vascular, diabetes):
    ### YOUR SOLUTION HERE
    """(int, str, boolean, boolean, boolean, boolean, boolean)->int
    This function takes 7 factors incluidng age in years, sex - Male/Female, 5 boolean variables 
    (CHF history, hypertension, stroek history, vascular disease history, and disbetes history), and 
    calculates the CHA2DS2-VASc Score for Atrial Fibrillation Stroke Risk. 
   
    >>> cha2ds2_vasc(50, 'Male', False, True, False, True, True)     
    3
    
    >>> cha2ds2_vasc(70, 'Female', False, True, True, True, False)     
    6

    """
    score = 0
    if age >= 65:
        score += 1
    if age >= 75:
        score += 1
    if sex == "Female":
        score += 1
    score += int(chf) + int(hypertension) + int(stroke)*2 + int(vascular) + int(diabetes)
    return score
````

## Part 2: HEART Score

The HEART score is a predictor for major cardiac events.  It requires 5 high-level inputs:
* History (Slightly / Moderately / Highly suspicious)
* EKG (Normal / Non-specific repolarization disturbance / Significant ST deviation)
* Age (Number)
* Risk Factors (Number of risk factors)
* Initial Troponin (Number of times the normal limit)

Fill out the function below with logic to calculate the numeric risk score for teh given input.

Be sure to provide meaningful documentation and at least two test cases in your documentation. Also make sure your code satisfies the test cases provided in the assert statements.

````python
def heart(history, ekg, age, risks, troponin):
    ### YOUR SOLUTION HERE
    """(str, str, int, int, float)->int
    This function caculates HEART score based on 5 factors: 
    - history (Slightly / Moderately / Highly suspicious)
    - EKG (Normal / Non-specific repolarization / Significant ST deviation)
    - age (# of years)
    - number of risk factors
    - initial troponin (3 of times of normal limit)
    
    >>> heart('Slightly suspicious','Normal',44,0,0.9)
    0
    
    >>> heart('Highly suspicious','Significant ST deviation',75,3,3.1)
    10
    """
    score = 0
    history_dict = {'Slightly suspicious':0, 'Moderately suspicious':1, 'Highly suspicious':2}
    EKG_dict = {'Normal':0, 'Non-specific repolarization':1, 'Significant ST deviation':2}
    score += history_dict[history]
    score += EKG_dict[ekg]
    if age >= 45:
        score += 1
    if age >= 65:
        score += 1
    if risks >= 1:
        score += 1
    if risks >= 3:
        score += 1
    if troponin > 1:
        score += 1
    if troponin > 3:
        score +=1
    return score        
````

## Part 3: Framingham Risk Score for Hard Coronary Heart Disease

The Framingham Risk Score for Hard Coronary Heart Disease is intended for non diabetic patients age 30-79 only.  So, if the patient's age is < 30 or > 79, your function should return `-1` rather than a specific risk score.

The Framingham Risk Score takes 7 inputs:
* Age (Number)
* Sex (Male / Female)
* Smoker (True / False)
* Total cholesterol (Number)
* HDL cholesterol (Number)
* Systolic BP (Number)
* Blood pressure being treated with medicines (No / Yes)

You'll not that rather than being a basic parametric equation, this is a regression function defined by coefficients.  It also requires you take the natural logarithm (`ln`) of many of the parameters.  To help you out, here's an example of how to interpret the formulat provided in the website's **Evidence** tab.

Take special note of the footnotes in the logic:
* *Yes=1, No=0 (for Treated for blood pressure and Smoker)
* ** Men: if age >70, use ln(70) x Smoker. Women: if age >78, use ln(78) x Smoker.


---

These segments of the equation and the coefficient table...

> $ L_{Men} = \beta \times \ln(Age) + \beta \times \ln(cholesterol) ... $
>
> | Variable &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; | Men | Women
> | :---------------------------- | :--: | :--: |
> | $ \ln(Age) $ | 52.00961 | 31.764001
> | $ \ln(Cholesterol) $ | 20.014077 | 22.465206

Could be written in Python as follows:

```python
import math

L = 0
if sex == "Male":
    L += 52.00961 * math.ln(age)
    L += 20.014077 * math.ln(cholesterol)

```

**ROUND YOUR FINAL RESULT TO 4 DECIMAL PLACES**

````python
def framingham(age, sex, smoker, cholesterol, hdl, systolic, bp_treated):
    ### YOUR SOLUTION HERE
    """(int, str, boolean, int, int, int, boolean)->float
    This function caculates Framingham Risk Score based on 7 factors: 
    - Age (Number)
    - Sex (Male / Female)
    - Smoker (True / False)
    - Total cholesterol (Number)
    - HDL cholesterol (Number)
    - Systolic BP (Number)
    - Whether blood pressure being treated with medicines (True / False)
    It returns -1 for patient age < 30 or >79.
    
    >>> framingham(60,'Female',True,156,63,124,True)
    0.0293
    
    >>> framingham(29,'Male',True,120,60,115,True)
    -1
    """
    import math
    L = 0
    if age < 30 or age > 79:
        P = -1
    else:
        if sex == "Male":
            L += 52.00961 * math.log(age)
            L += 20.014077 * math.log(cholesterol)
            L += -0.905964 * math.log(hdl)
            L += 1.305784 * math.log(systolic)
            L += 0.241549 * int(bp_treated)
            L += 12.096316 * int(smoker)
            L += -4.605038 * math.log(age) * math.log(cholesterol)
            L += -2.93323 * math.log(age) * math.log(age)
            L += -172.300168
            if age >70:
                L += -2.84367 * math.log(70) * int(smoker)
            else:
                L += -2.84367 * math.log(age) * int(smoker)
            P = round((1 - 0.9402 ** math.exp(L)), 4)
        else:
            L += 31.764001 * math.log(age)
            L += 22.465206 * math.log(cholesterol)
            L += -1.187731 * math.log(hdl)
            L += 2.552905 * math.log(systolic)
            L += 0.420251 * int(bp_treated)
            L += 13.07543 * int(smoker)
            L += -5.060998 * math.log(age) * math.log(cholesterol)
            L += -146.5933061
            if age >78:
                L += -2.996945 * math.log(78) * int(smoker)
            else:
                L += -2.996945 * math.log(age) * int(smoker)
            P = round((1 - 0.98767 ** math.exp(L)), 4)            
    return P     
            
            
````

First, you'll notice that some of the column names are different between the data in the input file and the values that are expected by your functions above.  For example: "M" from the file needs to be turned into "Male" and "Yes" in the file needs to be turned into "True".  You will need to do conversions for all of the fields listed below:

After calculating these three risk scores, use the rules below to determine who is at highest risk.  To be classified as "High Risk" a patient must meet all three criteria below:
1. CHA2DS2_VASc >= 2
2. HEART >= 4
3. Framingham >= 3%

Your output for this function needs to be a list where each item in this contains `[patient, CHA2DS2_VASc, HEART, Framingham, High Risk]`

````python
def test_patients(filename):
    ### YOUR SOLUTION HERE
    """(str)->list
    This function takes a string including the path and name of a csv file that 
    includes risk factors of invidivudal patients and returns a list including 
    each patient's ID, CHA2DS2_VASc score, HEART score, Framingham score, and whether 
    the patient has high overall risk.
    
    """
    
    import csv
    output = []
    with open(filename) as f:
        next(f)
        csv_reader = csv.reader(f)
        for row in csv_reader:
            patient = row[0]
            age = int(row[1])
            if row[2] == 'M':
                sex = 'Male'
            else:
                sex = 'Female'
            if row[3] == 'Yes':
                chf = True
            else:
                chf = False
            if row[4] == 'Yes':
                hypertension = True
            else:
                hypertension = False
            if row[5] == 'Yes':
                stroke = True
            else:
                stroke = False
            if row[6] == 'Yes':
                vascular = True
            else:
                vascular = False
            if row[7] == 'Yes':
                diabetes = True
            else:
                diabetes = False
            history = row[8]
            ekg = row[9]
            risks = int(row[10])
            troponin = float(row[11])
            if row[12] == 'Yes':
                smoker = True
            else:
                smoker = False
            cholesterol = int(row[13])
            hdl = int(row[14]) 
            systolic = int(row[15])
            if row[16] == 'Yes':
                bp_treated = True
            else:
                bp_treated = False
            CHA2DS2_VASc = cha2ds2_vasc(age, sex, chf, hypertension, stroke, vascular, diabetes)
            HEART = heart(history, ekg, age, risks, troponin)
            Framingham = framingham(age, sex, smoker, cholesterol, hdl, systolic, bp_treated)
            if CHA2DS2_VASc >= 2 and HEART >= 4 and Framingham >= 0.03:
                risk = True
            else: 
                risk = False
            output.append([patient, CHA2DS2_VASc, HEART, Framingham, risk])
    return output
````
